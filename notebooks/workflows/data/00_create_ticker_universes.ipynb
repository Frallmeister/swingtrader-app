# Create Ticker Universes

This notebook is used to manually create or refresh ticker universe YAML files under `src/swingtrader/configs/universes`. The generated YAML files are curated, committed artifacts that downstream runtime code reads; universe generation itself is a local/bootstrap workflow, not normal application runtime behavior.
